```python
from numpy import *
import pandas as pd
df1 = pd.read_csv('Construction-Emp.txt', sep='\t', error_bad_lines=False)
df2 = pd.read_csv('Bank-Failures.txt', sep='\t', error_bad_lines=False)
```

```python
#need to convert dates to a datetime format.
df1['observation_date'] = pd.to_datetime(df1['observation_date'],format='%Y/%m/%d')
df2['Failure Date'] = pd.to_datetime(df2['Failure Date'],format='%m/%d/%Y')
```

```python
#The dates have been changed to datetime format.
#Now dropping date and leaving year and month only.

df1['observation_date'] = df1['observation_date'].dt.strftime('%Y/%m')

df2['Failure Date'] = df2['Failure Date'].dt.strftime('%Y/%m')
```

```python
#list number of bank failures in each month/year
dfnew = df2['Failure Date'].value_counts().reset_index()
dfnew.head()
```

```python
df1
```

```python
names = {
    "index": "observation_date"
}

for col in dfnew.columns:
    if col in names:
        dfnew = dfnew.rename(columns = {col:names[col]})
        
```

```python
dfnew
```

```python
names = {
    "Failure Date": "Failure Count"
}

for col in dfnew.columns:
    if col in names:
        dfnew = dfnew.rename(columns = {col:names[col]})
        
```

```python
dfnew
```

```python
df3 = pd.merge(df1, dfnew, how="left", on="observation_date")
```

```python
df3
```

```python
df3 = df3.dropna()
#dropping observations with missing data
```

```python
df3
```

```python
df3['Construction_Emp'].pct_change()
```

```python
!pip install statsmodels
from pandas import Series
from matplotlib import pyplot
from statsmodels.graphics.tsaplots import plot_acf

series = df3[['Failure Count']]
plot_acf(series,lags=50)
pyplot.show()
```

```python
#Here, I see high autocorrelation for many lags, the highest being the first two, so it implies that the error is mostly related to the errors in t-1 and t-2, but is related up to around t-25.
```

```python
from statsmodels.graphics.tsaplots import plot_pacf

series = df3[['Failure Count']]
plot_pacf(series, lags=50)
pyplot.show()
```

```python
# Here, I see a large spike at lag 1, decreasing after a few lags. This may indicate a moving average term in the data. 
# I could also interpret this a different way. There are significant correlations in the first two lags, which are then followed by lags that are mostly insignificant. This indicates an autoregressive term of order 2.
```
