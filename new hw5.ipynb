```python
from numpy import *
import pandas as pd
df = pd.read_csv('PSID.txt', sep='\t', error_bad_lines=False)
print list(df)
df.describe()
```

```python
df = df.drop(df[df.ER60017 == 999].index)
# dropping observations where age of head is either DK or NA.
```

```python
df = df.drop(df[df.ER62366 == 8].index)
df = df.drop(df[df.ER62366 == 9].index)
# dropping observations where head's health is DK, NA, or refused.
```

```python
df = df.drop(df[df.ER62506 == 8].index)
df = df.drop(df[df.ER62506 == 9].index)
df = df.drop(df[df.ER62506 == 0].index)
# dropping observations where health status before age 17 is DK, NA; refused, or Inap.
```

```python
df = df.drop(df[df.ER64612 == 8].index)
df = df.drop(df[df.ER64612 == 9].index)
# dropping observations where any member in family w/o insurance.
```

```python
df = df.drop(df[df.ER64810 == 9].index)
# dropping observations where race of head is DK, NA, or refused.
```

```python
df = df.drop(df[df.ER65459 == 99].index)
# dropping observations where head's completed education level is DK or NA.
```

```python
df.loc[df['ER64612'] == 1, 'ER64612'] = 1
df.loc[df['ER64612'] == 5, 'ER64612'] = 0
#creating an indicator that take the value 1 for no insurance and 0 otherwise
```

```python
df.loc[df['ER60018'] == 1, 'ER60018'] = 0
df.loc[df['ER60018'] == 2, 'ER60018'] = 1
#creating an indicator that takes the value 1 for female and 0 otherwise.
```

```python
df = df.drop(df[df.ER65349 < 0].index)
#dropping when income is negative
```

```python
df.describe()
```

```python
df["ER65349"]=df["ER65349"].apply(lambda x:math.log(x+1))
#log transformation on income
```

```python
names = {
    "ER60001": "Release_Num",
    "ER60002": "ID",
    "ER60017": "Age",
    "ER60018": "Male",
    "ER62366": "Health_Now",
    "ER62506": "Health_Young",
    "ER64612": "NoHealthIns",
    "ER64810": "Race",
    "ER65349": "Income",
    "ER65459": "Educ"
}

for col in df.columns:
    if col in names:
        df = df.rename(columns = {col:names[col]})
df.head()
#changing names of columns
```

```python
df.loc[df['Race'] == 1, 'Race'] = 1
df.loc[df['Race'] == 2, 'Race'] = 2
df.loc[df['Race'] == 3, 'Race'] = 3
df.loc[df['Race'] == 4, 'Race'] = 3
df.loc[df['Race'] == 5, 'Race'] = 3
df.loc[df['Race'] == 7, 'Race'] = 3
#race: 1=white, 2=black, 3=other
```

```python
df.loc[df['Health_Now'] == 1, 'Health_Now'] = 1
df.loc[df['Health_Now'] == 2, 'Health_Now'] = 2
df.loc[df['Health_Now'] == 3, 'Health_Now'] = 2
df.loc[df['Health_Now'] == 4, 'Health_Now'] = 3
df.loc[df['Health_Now'] == 5, 'Health_Now'] = 3
#Health_Now: 1=excellent, 2=very good and good, 3=fair and poor
```

```python
df.loc[df['Health_Young'] == 1, 'Health_Young'] = 1
df.loc[df['Health_Young'] == 2, 'Health_Young'] = 2
df.loc[df['Health_Young'] == 3, 'Health_Young'] = 2
df.loc[df['Health_Young'] == 4, 'Health_Young'] = 3
df.loc[df['Health_Young'] == 5, 'Health_Young'] = 3
#Health_Young: 1=excellent, 2=very good and good, 3=fair and poor
```
