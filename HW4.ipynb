```python
from numpy import *
import pandas as pd
df1 = pd.read_csv('IRS-Tax Filing.txt', sep='\t', error_bad_lines=False)
df2 = pd.read_csv('Zillow-Rents.txt', sep='\t', error_bad_lines=False)
```

```python
df3=pd.merge(df1,df2, on='Zip')
```

```python
df3.describe()
```

```python
df3["ZrentValue"]=df3["ZrentValue"].apply(lambda x:math.log(x+1))
```

```python
df3["NoReturns"]=df3["NoReturns"].apply(lambda x:math.log(x+1))
```

```python
df3["SingleReturns"]=df3["SingleReturns"].apply(lambda x:math.log(x+1))
```

```python
df3["TaxableIncome"]=df3["TaxableIncome"].apply(lambda x:math.log(x+1))
```

```python
df3["StudentLoanDeductions"]=df3["StudentLoanDeductions"].apply(lambda x:math.log(x+1))
```

```python
df3["UnempComp"]=df3["UnempComp"].apply(lambda x:math.log(x+1))
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
X, y = df3[['NoReturns','SingleReturns','TaxableIncome','StudentLoanDeductions','UnempComp']], df3.ZrentValue
model = lm.fit(X, y)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
lm.score(X, y)
```

```python
import statsmodels.api as sm
X = sm.add_constant (X)
model = sm.OLS(y,X).fit()
model.summary()
```

```python
# Brief discussion of model and estimation procedure

# I used a linear regression model to find out the 
# effect of the median rent value on the number of returns filed,
# number of single returns filed, total taxable income, 
# total student loan interest deductions, and total unemployment 
# compensation. Before running the regression, I did a log transformation
# so that my results would be less sensitive to outliers in the data. 
```

```python
#An analysis and discussion of the findings

#Because I did a log-log transformation, my coefficients reflect an 
#estimated percent change in rent for a percent change in the 
#independent variables. Looking at Unemployment Compensation: a 1% 
#increase in unemployment compensation is associated with a 0.0066% 
#increase in median rent value, on average, holding other variables
#constant. This increase in rent value is because unemployment 
#made homeowners switch to renting. The increase in demand for 
#rentals drove up prices.

#The coefficient for single returns is positive because
#single individuals are more likely to rent in an area.
#more demand for rent means higher rent values. 

#Looking at the p-values, number of returns filed and 
#total student loan interest deductions seem to be statistically 
#insignificant. 
```
