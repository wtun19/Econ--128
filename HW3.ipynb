```python
#Question 2: On the x-axis of this graph are the dates from 2000 to 2017. 
#On the y-axis are the average daily difference between closing and opening prices.
#For the first two thirds of the graph (approximately 2000 to 2011), the average daily
#difference is very low. There is almost no volatility. This means stock prices are quite 
#stable throughout the day. However from around 2011 to 2017, there is much more 
#volatility. The extreme peaks show that price increased by a lot on those days, 
#and dips in the graph shows that price decreased by a lot on those days. Even when
#not taking these extremes into account, overall, 2011 to 2017 shows a greater volatility
#in daily price compared to 2000-2011.
```

```python
#Question 3 starts here.
```

```python
df['Difference2'] = df['High'] - df['Low']
```

```python
df['Daily Price Variation'] = df['Difference2'] / df['Close']
```

```python
df.groupby(['Date'])['Daily Price Variation'].mean()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

groups = df.groupby(['Date'])['Daily Price Variation'].mean()
plt.axis([1975,2017,0,0.15])
groups.plot.bar()
```

```python
#Question 3: Please note that the bar graph does take a while to load.
```

```python
# Daily price variation measures the stockâ€™s volatility, 
# or how much its value changes. The values on the X-axis are 
# the dates from 1975 to 2017, and the values on the Y-axis are
# the average daily price variations. Low average daily price 
# variations means that the value of the stock in the market is stable,
# while high average daily price variations means the stockâ€™s value 
# is very unpredictable. We can see from the graph that over time, 
# the average daily price variations somewhat resemble a dampened sine 
# wave. Average daily price variations generally oscillate between 
# increasing and decreasing, but each successive peak (highest point) 
# decreases over time. This means stock prices are becoming less volatile
# as time goes on. You can see on the right side of the graph that over the
# past few years, average daily price variations are at their lowest and most stable. 
```
